```python
# 猜数字

import random

def guess_number(max_value=10):
    target = random.randint(1, max_value)
    print(f"猜数字游戏开始啦~~我已经想好了一个 1 ~~ {max_value} 的是数字~~")

    num = 0
    while num < 3:
        try:
            guess = int(input("来猜一猜："))
        except ValueError:
            print("必须输入数字！")
            continue
        if guess < target:
            print("小了小了~~再往大一点试试？")
        elif guess > target:
            print("大了大了~~往小了猜猜？")
        else:
            print("恭喜你，才对啦！")
            break
        num += 1
    else:
        print(f"游戏失败，正确答案是{target}")

guess_number(20)
```
