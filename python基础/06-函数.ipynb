```python
# 自定义函数
def my_sum(a, b):
    res = a + b
    return res


print(my_sum(1, 2))
```

```python
def change(a, b):
    return b, a

res = change(1, 2)
print(res, type(res))
a, b = change(3, 4)
print(a, b)
```
