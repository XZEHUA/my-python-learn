```python
# 必填参数（位置参数）
def demo(name, age):
    print(name, age)

demo('fei', 18)
```

```python
# 缺省参数（默认参数）
def demo(name, age, city='成都'):
    print(name, age, city, sep="-", end="。")

demo('fei', 18)
demo('fei', 18, '上海')
```

```python
# 可变参数：参数的数量不受限制
def add(*num):
    print(num, type(num))
    total = 0
    for n in num:
        total += n      # total = total + n
    return total


print(add(1, 2, 3))

# 可变参数的序列解包功能
l = [1,2,3]
add(*l)
```

```python
# 关键字参数
def demo(name, age, city='成都', **other):
    print(name, age, city)
    print(other, type(other))

demo('fei', 18, '上海', gender = 'male', tel = 10086)
d = {"gender":"male", "tel":10086}
demo('fei', 18, '上海',**d)
```
