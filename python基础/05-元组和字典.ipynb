```python
# tuple 元组，和 list 相同, 只读且用小括号表示
t = ('python', 'java', 1 ,2, 3, 4, 5)
print(t, type(t))
print(t[:3])
del t
print(t)
```

```python
t1 = (1, 2, 3)
print(t1, id(t1))
t2 = t1 + (4, 5, 6)
print(t1, id(t1))
print(t2, id(t2))
```

```python
# dict 字典
d = {'name': 'fei', 'age': 18, 'gender': 'male'}
print(d['name'], d.get('city', 'chengdu'))
for item in d:
    print(item)

for item in d.values():
    print(item)

for item in d.items():
    print(item)

for key, value in d.items():
    print(key, value)


# 增
d["country"] = 'china'
print(d)

# 改
d['age'] = 19
print(d)

# 删
del d['gender']

print(d)
```

```python
s = {}      # 默认是字典
print(type(s))

s = set()   # 创建空集合
print(type(s))

t = (1, 2, 3)
print(type(t))

t = (1,)     # 如果元组只有一个元素，后面必须添加逗号
print(type(t))
```
