```python
# 变量
a = 10  # 变量的类型取决于数据
print(type(a))
a = "10"
print(type(a))
```

```python
# 字符串类型
s = "hello python"
print(s[0])

# 切片
print(s[0:5])   # 左闭右开区间
print(s[6:])
print(s[:5])
print(s[6:-1])
```

```python
# 字符串的格式化输出
d = 3.1415926
print(f"保留两位小数：{d:.2f}")
print(f"保留两位小数：{d:010.2f}")
```
