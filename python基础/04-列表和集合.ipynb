```python
# list 列表
l = [10, "abc", False, 3.14, 'python', "python"]
print(l, type(l), len(l))

# 因为列表有顺序，所以可以追加元素
l.append('java')

# 在指定下标插入元素
l.insert(1, 'java')
print(l)

# 删除
print(l.pop(-2))
print(l.remove('abc'))
del l[1]
print(l)

# 修改
l[0] = 'js'


# 查
print(l[0])
print(l.index('python'))
for i in l:     # 序列，可以被拆分的元素：str list set tuple dict 可迭代对象
    print(i)
```

```python
# set 集合
s = {'a', 'a', 'B', 'b', 1}
print(s, type(s))
s.add("python")
print(s.remove('a'))
print(s.pop())
print(s)
for i in s:
    print(i)
```
