```python
# while 循环一般用于不知道循环次数的场景
while True:
    email = input("请输入邮箱：")
    index = email.find("@")
    if index == -1:
        print("邮箱格式不正确！")
    elif index == 0:
        print("不能以@开头！")
    elif index == len(email)-1:
        print("不能以@结尾！")
    else:
        print(f"你的邮箱用户名为：{email[:index]}, 你的邮箱类型为：{email[index+1:]}")
        break
```

```python
# for 循环，适合明确知道循环次数的场景
for i in "python":
    print(i)
```

```python
for i in range(1,5):
    if i == 2:
        continue
    print(i)
    if i == 3:
        break
```

```python
for i in range(3):
    email = input("请输入邮箱：")
    index = email.find("@")
    if index == -1:
        print("邮箱格式不正确！")
    elif index == 0:
        print("不能以@开头！")
    elif index == len(email)-1:
        print("不能以@结尾！")
    else:
        print(f"你的邮箱用户名为：{email[:index]}, 你的邮箱类型为：{email[index+1:]}")
        break
else:
    print(f"你已连续输错{i+1}次！")
```
