```python
email = input("请输入邮箱：")
index = email.find("@")
if index == -1:
    print("邮箱格式不正确！")
elif index == 0:
    print("不能以@开头！")
elif index == len(email)-1:
    print("不能以@结尾！")
else:
    print(f"你的邮箱用户名为：{email[:index]}, 你的邮箱类型为：{email[index+1:]}")
```

```python
# if 嵌套
num = int(input("请输入一个数字："))
# print(type(num))
if num % 2 == 0:
    if num % 3 == 0:
        print("你输入的数字可以被2和3整除。")
    else:
        print("你输入的数字可以被2整除。")
else:
    if num % 3 == 0:
        print("你输入的数字可以被3整除。")
    else:
        print("你输入的数字不可以被2和3整除。")
```
